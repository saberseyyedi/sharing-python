## Sub Plotting

```python
x = np.linspace(0,100)
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)
```

```python
fig = plt.figure(figsize=(10, 7))
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Test-1") # add title to a subplot
plt.xlabel('This is X')
plt.ylabel('This is Y1')
plt.grid(linestyle=':')
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Test-2") # add title to a subplot
plt.xlabel('This is X')
plt.ylabel('This is Y2')
plt.grid(linestyle=':')
plt.subplot(2, 2, 3)
plt.plot(x, y3);
plt.title("Test-3") # add title to a subplot
plt.xlabel('This is X')
plt.ylabel('This is Y3')
plt.grid(linestyle=':')
plt.subplot(2, 2, 4)
plt.plot(x, y4);
plt.title("Test-4") # add title to a subplot
plt.xlabel('This is X')
plt.ylabel('This is Y4')
plt.grid(linestyle=':')
plt.suptitle("This is Super Title"); # add super title
plt.tight_layout(pad=2) # add padding to subplots
```

```python
fig, ax = plt.subplots(3, 2, figsize=(10,12))
fig.suptitle("Some Plots")
ax[0,0].plot(x, y1)
ax[0,1].plot(x, y2)
ax[1,0].plot(x, y3)
ax[1,1].plot(x, y4)

ax[0,0].set_title("A")
ax[0,1].set_title("B")
ax[1,0].set_title("C")
ax[1,1].set_title("D")
ax[2,0].set_title("E")
ax[2,1].set_title("F")

ax[0,0].set_xlabel('x')
ax[0,1].set_xlabel('x')
ax[1,0].set_xlabel('x')
ax[1,1].set_xlabel('x')

ax[0,0].set_ylabel('$y={x}$')
ax[0,1].set_ylabel('$y={x}^2$')
ax[1,0].set_ylabel('$y={x}^3$')
ax[1,1].set_ylabel('$y=\sqrt[2]{x}$')


fig.tight_layout(pad=3.0)
plt.show()
```

# Histogram

Normal Form Formula
<img src="normal.png"/> 


```python
mu = 100
sigma = 15
x = np.random.normal(mu,sigma,(1000,))
```

```python
x.shape[0]
```

```python
plt.scatter(np.linspace(0,1, x.shape[0]), x, marker='.', s=10)
plt.plot([0,1],[100,100],'--r');
```

```python
def norm_form(mu, sigma, bins): 
    
    scale = (1 / (np.sqrt(2 * np.pi) * sigma))
    x1 = 1 / sigma * (bins - mu)
    y =  scale * np.exp(-0.5 * (x1)**2)
    return y
```

```python
num_bins = 50
fig, ax = plt.subplots()
n, bins, patches = ax.hist(x, num_bins, density=True) 

'''
    "n" is number elements in an especific bin
    "bins" is number of bins in the histogram
    "density" scales the histoglam to [0,1]. when the density is True then height
    is less than 1.0
'''

norm_form(mu, sigma, bins)

ax.plot(bins, y, '--')
ax.set_xlabel('X')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram: $\mu=100$, $\sigma=15$');

fig.tight_layout()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(121)
n, bins, patches = plt.hist(x, 'auto', density=False)

plt.subplot(122)
n, bins, patches = plt.hist(x, 'auto', density=True)

plt.show()
display(bins)
display(len(bins))
display(patches)
```

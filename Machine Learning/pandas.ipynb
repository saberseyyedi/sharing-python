```python
import numpy as np
import pandas as pd
```

## Use list to make data in a Row

```python
# By Row
pd.DataFrame([[1,2,3],
             [4,5,6],
             [7,8,9]])
```

## Use Dictionary to make data in a Column

```python
# By Column
df = pd.DataFrame({'Name': ['Saber','Nelin','Rayan','Sara'],
               'Height': [176,110,112,168]})
```

```python
df
```

```python
df.Name
```

```python
df['Name']
```

```python
# Use the Numpy Syntax
df.loc[1,'Name']
```

```python
df.loc[1,:]
```

```python
df.loc[:,'Name']
```

```python
df.iloc[2,0] # iloc assumes that Name is 0 and Height is 1
```

### <b style="color:red">Tip: </b>If there are values that are different from the default NaN values of Pandas we can add them to the NaN list using 'na_values'


```python
df = pd.read_csv('property data.csv', na_values=['--','na']) # add '--' and 'na' to NaN list
df
```

```python
df.dropna() # Eliminate the rows include NaN value
```

```python
df.info()
```

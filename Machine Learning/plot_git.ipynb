```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## plot & scatter

```python
x = np.array([1,2,3,4])
y = np.array([8,14,4,0])
```

```python
plt.plot(x,y);
```

```python
plt.figure(figsize=(10,3))
plt.subplot(131)
plt.plot(x,y,'s')
plt.subplot(132)
plt.plot(x,y,'o')
plt.subplot(133)
plt.plot(x,y,'.');
```

```python
plt.figure(figsize=(12,4))
plt.subplot(131)
plt.plot(x,y,'D-')
plt.subplot(132)
plt.plot(x,y,'H-')
plt.subplot(133)
plt.plot(x,y,'*-');
```

```python
plt.figure(figsize=(10,3))
plt.subplot(131)
plt.plot(x,y,'s--')
plt.subplot(132)
plt.plot(x,y,'o--')
plt.subplot(133)
plt.plot(x,y,'.--');
```

```python
plt.figure(figsize=(10,3))
plt.subplot(131)
plt.plot(x,y,'s:')
plt.subplot(132)
plt.plot(x,y,'o:')
plt.subplot(133)
plt.plot(x,y,'.:');
```

```python
plt.figure(figsize=(10,3))
plt.subplot(131)
plt.plot(x,y,'s-r')
plt.subplot(132)
plt.plot(x,y,'o-g')
plt.subplot(133)
plt.plot(x,y,'.-b');
```

```python
plt.figure(figsize=(10,2))
plt.subplot(141)
plt.plot(x,y,'^-c')
plt.subplot(142)
plt.plot(x,y,'v-m')
plt.subplot(143)
plt.plot(x,y,'<-y');
plt.subplot(144)
plt.plot(x,y,'>-k');
```

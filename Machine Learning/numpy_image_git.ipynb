# Convert color photos to black and white

```python
im9 = (im8[:,:,0]/3 + im8[:,:,1]/3 + im8[:,:,2]/3).astype('uint8') 
plt.imshow(im9, 'gray');
```

```python
im10 = im8.copy().astype('uint16')
im11 = ((im10[:,:,0] + im10[:,:,1] + im10[:,:,2])/3).astype('uint8') 
plt.imshow(im11, 'gray');
```

```python
im12 = np.mean(im8, axis=2) 
plt.imshow(im12, 'gray');
```

# Luminance
- I = 0.2126 * R + 0.7152 * G + 0.0722 * B

```python
im13 = (im8[:,:,0]*0.2126/3 + im8[:,:,1]*0.7152/3 + im8[:,:,2]*0.0722/3).astype('uint8') 
plt.imshow(im13, 'gray');
```

```python
M = np.array([0.2126,0.7152,0.0722]).reshape(3,1)
print(im8.shape)
print(M.shape)
```

```python
im8.dot(M)
```

```python
im8.dot(M).shape
```

```python
plt.imshow(im8.dot(M),'gray');
```
